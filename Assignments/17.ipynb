```python
# Find the sum of odd and even series using functions

import math

def odd_series(n):
    s = 0
    for i in range(1, n + 1, 2):
        s += (i ** 2) / math.factorial(i)
    return s

def even_series(n):
    s = 0
    for i in range(2, n + 1, 2):
        s += (i ** 2) / math.factorial(i)
    return s

n = int(input("Enter n: "))

print("Sum of Odd Series =", odd_series(n))
print("Sum of Even Series =", even_series(n))
```
