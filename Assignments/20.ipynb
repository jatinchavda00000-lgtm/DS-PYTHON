```python
# Mini Project: Password Generator

import random
import string

class User:
    def __init__(self, user_id, name, password):
        self.details = (user_id, name, password)

    def display(self):
        print("User Details:", self.details)

try:
    user_id = int(input("Enter User ID: "))
    name = input("Enter Name: ")
    text = input("Enter some words: ")

    words = text.split()

    if len(words) < 2:
        raise ValueError("Please enter at least two words.")

    # Select random words
    password = random.choice(words).capitalize()
    password += random.choice(words)

    # Add numbers and special characters
    password += str(random.randint(10, 99))
    password += random.choice("!@#$%^&*")

    # Ensure length > 8
    while len(password) <= 8:
        password += random.choice(string.ascii_letters)

    print("Generated Password:", password)

    user = User(user_id, name, password)
    user.display()

except ValueError as e:
    print("Error:", e)

except Exception as e:
    print("Unexpected Error:", e)
```
